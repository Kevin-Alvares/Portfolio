# **Problem Statement:**
### The objective is to identify and create distinct customer profiles for each type of treadmill offered by an Indian fitness equipment brand. The brand offers three different types of treadmills and aims to understand the specific characteristics of their target audience for each treadmill model. This will facilitate the improvement of recommendations for new customers, ensuring that they are matched with the treadmill model that best suits their preferences and needs.

### To achieve this goal we will
### 1) Develop distinct customer profiles for each type of treadmill product through descriptive analytics, utilizing relevant tables and charts.
### 2) Construct two-way contingency tables to calculate conditional and marginal probabilities, uncovering insights into customer preferences and their impact on treadmill choices.

#Conditional probability

```python
event_A = 'KP281'  # Purchasing Product A
event_B = "Single"    # Fitness Level 4

# Step 2: Count the Occurrences
# Count the number of customers who purchased Product A and have Fitness Level 4
count_A_and_B = data[(data['Product'] == event_A) & (data['MaritalStatus'] == event_B)].shape[0]

# Count the number of customers who have Fitness Level 4
count_B = data[data['MaritalStatus'] == event_B].shape[0]

# Step 3: Calculate the Conditional Probability
conditional_probability = count_A_and_B / count_B

# Print the result
print(f"Conditional Probability of purchasing Product {event_A} given Fitness Level {event_B}: {conditional_probability:.2f}")
```
